```python

class comparingMethods:
    @staticmethod
    def compare_Seed(teamA, teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'Seed'] > df.at[teamB_indicie, 'Seed']):
            return 6
        else:
            return -6




    def compare_Wins(teamA, teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'Wins'] > df.at[teamB_indicie, 'Wins']):
            return 4
        else:
            return -4
    
    @staticmethod
    def compare_RPI(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'TeamRPI'] > df.at[teamB_indicie, 'TeamRPI']):
            return 7
        else:
            return -7

    @staticmethod
    def compare_RPIrank(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'TeamRPIRank'] > df.at[teamB_indicie, 'TeamRPIRank']):
            return -9
        else:
            return 9

    @staticmethod
    def compare_strengthofschedule(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'StrenghOfSchedule'] > df.at[teamB_indicie, 'StrenghOfSchedule']):
            return 4
        else:
            return -4

    @staticmethod
    def compare_fieldgoalpercent(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'FieldGoalPercent'] > df.at[teamB_indicie, 'FieldGoalPercent']):
            return 4
        else:
            return -4

    @staticmethod
    def compare_PPG(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'PtsPerGame'] > df.at[teamB_indicie, 'PtsPerGame']):
            return 4
        else:
            return -4

    @staticmethod
    def compare_3ptpercent(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, '3PTPercent'] > df.at[teamB_indicie, '3PTPercent']):
            return 3
        else:
            return -3
        
    @staticmethod
    def compare_freethrowpercent(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'FreeThrowPercent'] > df.at[teamB_indicie, 'FreeThrowPercent']):
            return 2
        else:
            return -2

    @staticmethod
    def compare_APG(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'AssistsPerGame'] > df.at[teamB_indicie, 'AssistsPerGame']):
            return 3
        else:
            return -3

    @staticmethod
    def compare_TPG(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'TurnoversPerGame'] > df.at[teamB_indicie, 'TurnoversPerGame']):
            return -4
        else:
            return 4

    @staticmethod
    def compare_APGtoTPG(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'AssistToTurnoverRatio'] > df.at[teamB_indicie, 'AssistToTurnoverRatio']):
            return 7
        else:
            return -7

    @staticmethod
    def compare_PPGagainst(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'PtsAllwdPerGame'] > df.at[teamB_indicie, 'PtsAllwdPerGame']):
            return 1
        else:
            return -1

    @staticmethod
    def compare_PPGdiff(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'PtDifferential'] > df.at[teamB_indicie, 'PtDifferential']):
            return 7
        else:
            return -7

    @staticmethod
    def compare_RBG(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'RebPerGame'] > df.at[teamB_indicie, 'RebPerGame']):
            return 3
        else:
            return -3

    @staticmethod
    def compare_RBGagainst(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'RebAllwdPerGame'] > df.at[teamB_indicie, 'RebAllwdPerGame']):
            return -1
        else:
            return 1

    @staticmethod
    def compare_offRBG(teamA,teamB):
        teamA_indicie =  df.index[df['Team'] == teamA].tolist()[0]
        teamB_indicie =  df.index[df['Team'] == teamB].tolist()[0]
        if(df.at[teamA_indicie, 'OffRebPerGme'] > df.at[teamB_indicie, 'OffRebPerGme']):
            return 3
        else:
            return -3
```
